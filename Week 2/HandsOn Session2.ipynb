# Filtering and Sorting

```python
drinks.head(10)
```

```python
eu_popular_beers = drinks[(drinks['beer_servings'] > 200) & (drinks['continent'] == 'EU')]
eu_popular_beers
```

```python
eu_popular_beers.sort_values(by=['beer_servings', 'total_litres_of_pure_alcohol'], ascending=False)
```

```python
eu_beers = drinks[drinks['continent'] == 'EU']
eu_beers.describe(include="all")
```

```python
eu_a = eu_beers[(eu_beers['beer_servings'] > 193.778)]
eu_a.describe()
```

```python
eu_beers[eu_beers['total_litres_of_pure_alcohol'] == 14.400]
```

```python
drinks['servings'] = drinks.beer_servings + drinks.spirit_servings + drinks.wine_servings
drinks['ml'] = drinks.total_litres_of_pure_alcohol*1000
```

```python
drinks.head(5)
```

```python
drinks.drop('ml', axis=1, inplace=True)
drinks.drop('servings', axis=1, inplace=True)
```

```python
drinks.head(2)
```

```python
drinks[drinks['continent'].isnull()]
```

```python
drinks.isnull().sum(axis=0)
```

```python
#Fill null values with some other values
drinks['continent'].fillna(value='NA', inplace=True)
drinks.isnull().sum(axis=0)
```

```python
drinks['total_litres_of_pure_alcohol'].max(axis=0)
```

```python
drinks.groupby('continent').beer_servings.mean()
```

```python
drinks.groupby('continent').beer_servings.agg(['count', 'mean', 'min', 'max'])
```
